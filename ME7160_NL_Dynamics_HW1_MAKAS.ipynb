# Problem 1A

$\dot{x}+x=0$, for $x(0)=5$

Assume solution of the form $x(t)=e^{rt}$, therefore now we get
$(r+1)=0$, which yields $r=-1$.

Therefore solution becomes $x(t)=Ce^{-t}$

After applying B.C. $x(0)=5$ final result becomes:  

### **$x(t)=5e^{-t}$**


```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt

#Define time interval and calculate x(t)
t=np.linspace(0.,10.,1000)
x=5*np.exp(-t)

#Plot the results
fig=plt.figure()
plt.plot(t,x)
#pylab.xlim([0, 8])
#pylab.ylim([-1,6])
plt.axis([0, 8, -1, 6])
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

# Problem 1B

$10\ddot{x}+\dot{x}+100x=0$ for $x(0)=5$ and $\dot{x}(0)=2$

After rearranging terms equation can be written as:
$\ddot{x}+\frac{1}{10}\dot{x}+10x=0$

From this form we can get that $\omega_n=\sqrt{10}$ and $2\epsilon\omega_n=\frac{1}{10}$, which yields $\epsilon=0.0158$
For this $2^{nd}$ order ODE roots come in the form $r_{1,2}=-\epsilon\omega_n\pm\omega_n\sqrt{\epsilon^{2}-1}$. Since $\epsilon<1$ the solution is underdamped and will take the form of:

$x(t)=Ae^{-\lambda t}*cos(\mu t)+Be^{-\lambda t}*sin(\mu t)$

where $\lambda=-\epsilon\omega_n=-0.05$ & $\mu=\omega_n\sqrt{\epsilon^2-1}=3.16$

After applying B.C's $A=5$ and $B=0.7116$, putting everything together gives:

### **$x(t)=5e^{-0.05 t}*cos(3.16 t)+0.7116e^{-0.05 t}*sin(3.16 t)$**

Time Constant $\tau=\frac{1}{\epsilon\omega_n}=20$

Settling Time (in 4 time constants) $T_s=\frac{4}{\epsilon\omega_n}=80.0$

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt

#Define time interval and calculate x(t)
t=np.linspace(0.,100.,2000)
A=5
B=0.7116
y=-0.05
u=3.16
x=A*np.exp(y*t)*np.cos(u*t)+B*np.exp(y*t)*np.sin(u*t)

#Plot the results
fig=plt.figure()
plt.plot(t,x)
#pylab.xlim([0, 8])
#pylab.ylim([-1,6])
plt.axis([0, 100, -6, 6])
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

# Problem 1C

Rearrange the equation to obtain: $\frac{\dot{x}}{x-\frac{1}{5}}=-10$, from this expression it is evident that left hand side is a derivative of $ln|x-\frac{1}{5}|$, Therefore now we have $\frac{d}{dt}ln|x-\frac{1}{5}|=-10$.

### $x=\frac{1}{5}+e^{-10t-0.223}$

# Problem 1D  

$10\ddot{x}+2\dot{x}+100x=200$, where $x(0)=0$, and $\dot{x}(0)=0$  

First using above expression solve for particular solution, assume $X(t)=2A$. Plugging in $X(t)$ into the ODE gives that $A=1$ so $X(t)=2$.  

Next rearrange above ODE to get $\ddot{x}+\frac{1}{5}\dot{x}+10x=2$. From this expression $\omega_n=\sqrt{10}$ and $2\epsilon\omega_n=\frac{1}{5}$, from which $\epsilon=0.0316$.  

Next solve for the characteristic solution, which due to the fact that $\epsilon<1$ will be of the form:  

$x(t)=Ae^{-\lambda t}*cos(\mu t)+Be^{-\lambda t}*sin(\mu t)+2$

where $\lambda=-\epsilon\omega_n=-0.10$ & $\mu=\omega_n\sqrt{\epsilon^2-1}=3.16$  

After applying B.C's $A=-2$ and $B=0.0633$, putting everything together gives:

### **$x(t)=-2e^{-0.10 t}*cos(3.16 t)+0.0633e^{-0.10 t}*sin(3.16 t)+2$**  

Time Constant $\tau=\frac{1}{\epsilon\omega_n}=10$

Settling Time (in 4 time constants) $T_s=\frac{4}{\epsilon\omega_n}=40.0$

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt

#Define time interval and calculate x(t)
t=np.linspace(0.,100.,2000)
A=-2
B=0.0633
y=-0.10
u=3.16
x=A*np.exp(y*t)*np.cos(u*t)+B*np.exp(y*t)*np.sin(u*t)+2

#Plot the results
fig=plt.figure()
plt.plot(t,x)
#pylab.xlim([0, 8])
#pylab.ylim([-1,6])
plt.axis([0, 50, 0, 4])
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

# Problem 2A  

$\dot{x}+10x=0$  

First plot the free response for the above expression.

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def deriv(x1,t):
    return (-10*x1)

time=np.linspace(0.0,10.0,1000)
yinit=1
x1=odeint(deriv, yinit, time)

fig=plt.figure()
plt.plot(time,x1)
plt.axis([0, 1, -0.5, 1])
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

The roots of the equation are ploted below. Since the solution converges assymptotically. It is expected that the root should lie on the left plane, which is the case below.

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#solve for roots
poly = [1, 10]
a=np.roots(poly)

#plot roots
fig=plt.figure()
plt.plot(a.real,a.imag,'ro-')
plt.axis([-15, 5, -4, 4])
fig.suptitle('Plot Re versus Im', fontsize=15)
plt.xlabel('Re')
plt.ylabel('Im')
plt.grid()
plt.show()
```

# Problem 3B  

$10\ddot{x}+\dot{x}+100x=0$, rearrange to get $\ddot{x}+\frac{1}{10}\dot{x}+10x=0$  

Now plot the free response and phase plot.  

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def deriv(x,t):
    return np.array([x[1], -10*x[0]-(1/10)*x[1]])

time=np.linspace(0.0,100,2000)
xinit=np.array([1, 1])
x=odeint(deriv, xinit, time)

#Plot free response results
fig=plt.figure()
plt.plot(time,x[:,0])
plt.axis([0, 100, -1.2, 1.2])
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')
plt.grid()
plt.show()

#Plot phase plot
fig1=plt.figure()
plt.plot(x[:,1],x[:,0])
plt.axis([-4, 4, -1.2, 1.2])
fig1.suptitle('Phase plot for x(t)', fontsize=15)
plt.xlabel('v')
plt.ylabel('x')
plt.grid()
plt.show()
```

The phase plot shows spiral stability.

# Problem 3C  

$\dot{x}+x-\frac{1}{2}x^{3}=0$, for this function set $\dot{x}=0$ to find equilibrium points.  

After rearranging the equation we get $x(1-\frac{1}{2}x^{2})=0$, from this the equilibrium points are $x=0$ and $x=\pm\sqrt{2}$. If the initial condition is within the bounds of $\pm\sqrt{2}$ the solution will converge.

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from math import sqrt

def deriv(x1,t):
    return (-x1+(1/2)*(x1**3))

time=np.linspace(0.0,1.0,1000)
xinit=sqrt(2)
x1=odeint(deriv, xinit, time)

fig=plt.figure()
plt.plot(time,x1)
fig.suptitle('Plot x(t) versus time(t)', fontsize=15)
plt.xlabel('Time, t')
plt.ylabel('x(t)')

def deriv(x2,t):
    return (-x2+(1/2)*(x2**3))

time1=np.linspace(0.0,1.0,1000)
xinit=-sqrt(2)
x2=odeint(deriv, xinit, time)

plt.plot(time,x2)
plt.axis([0, 1, -1.5, 1.5])
plt.grid()
plt.show()
```

Next phase plot is generated with above limits in mind using equation $\dot{x}=-x+\frac{1}{2}x^{3}$.

```python
%matplotlib inline

import pylab
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

#Define time interval and calculate x(t) for stable region
x=np.linspace(-sqrt(2.0),sqrt(2.0),100)
xdot=-x+(1/2)*(x**3)

#Define unstable region in negative plane
x1=np.linspace(-sqrt(2.0),-2,100)
xdot1=-x1+(1/2)*(x1**3)

#Define unstable regoin in positive plane
x2=np.linspace(sqrt(2.0),2,100)
xdot2=-x2+(1/2)*(x2**3)

#Plot the results
fig=plt.figure()
plt.plot(xdot,x)
plt.plot(xdot1,x1,'r--')
plt.plot(xdot2,x2,'r--')
#plt.axis([0, 8, -1, 6])
fig.suptitle('Phase Plot for x(t)', fontsize=15)
plt.xlabel('v')
plt.ylabel('x')
plt.grid()
plt.show()
```

The blue region represents the stable region where as the red region is the area of instability. The solution will be assymptotically stable as long as it stays within bounds.

# Problem 4  

$\ddot{x}=-\mu(x^{2}-1)\dot{x}-x$, after rearranging the equation $\ddot{x}+\mu(x^{2}-1)\dot{x}+x=0$ it is evident that $\mu$ is acting as a damping coefficient.  



```python
%matplotlib inline
 
import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def prob4(u=0, t1=100.0):
    
    def deriv(x,t):
        #u=0
        return np.array([x[1], -u*(x[0]**2-1)*x[1]-x[0]])

    time=np.linspace(0.0,t1,2000)
    xinit=np.array([1, 1])
    x=odeint(deriv, xinit, time)
    
    #Plot free response results
    fig=plt.figure()
    plt.plot(time,x[:,0])
    #plt.axis([0, 100, -1.2, 1.2])
    fig.suptitle('Plot x(t) versus time(t) at u=' + str(u) + ' and Time =' + str(t1) + ' sec', fontsize=15)
    plt.xlabel('Time, t')
    plt.ylabel('x(t)')
    plt.grid()
    plt.show()

    #Plot phase plot
    fig1=plt.figure()
    plt.plot(x[:,1],x[:,0])
    #plt.axis([-4, 4, -1.2, 1.2])
    fig1.suptitle('Phase plot for x(t) at u=' + str(u) + ' and Time =' + str(t1) + ' sec', fontsize=15)
    plt.xlabel('v')
    plt.ylabel('x')
    plt.grid()
    plt.show()
```

```python
prob4(u=-1)
```

With $\mu<0$ the system settles quickly. The phase plot shows spiral stability

```python
prob4(u=0)
```

With $\mu=0$ the system has no damping and behaves marginally stable. Phase plot shows center stability.

```python
prob4(u=.5)
```

With $\mu>0$ the amplitude increases until it reaches a value of 2 then levels out. Phase plot shows center stability.

```python
prob4(u=2, t1=100)
```

With $\mu$ increasing the free response, while still sinosoidal, begins to form peaks at top of the response. The reponse still levels off at 2 but it gets there in quicker time.

```python
prob4(u=5)
```

with further increase in $\mu$ the sharp peaks begin to be more prominant. The response still appears to be center stable based on phase plot.

```python
prob4(u=10, t1=100)
```

With $\mu=10$ the peaks continue to become more pronounced. At this $\mu$ value it appears that the frequency of the oscillations seems to be increasing with time. The response still levels off at a value of 2. The free response is also exhibiting near infinite $\dot{x}$. As a result the phase plot is beginning to experience sharp directional changes.  

**In general points on the inside of the graph travel outward until they meet up with the graph. In similar fashion points on the outside of the graph travel inward untill they meet up with the graph.**
